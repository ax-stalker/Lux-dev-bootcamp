```python
# imports
import pandas as pd
```

```python

df =pd.read_csv("weather_data.csv")
df.head()
```

```python
# records where the weather was exactly clear.
clear_mask= df["Weather"] == "Clear"

df[clear_mask].head()

```

```python
# Times where windspeed was exactly 4km/hr
four_mask =df['Wind Speed_km/h']== 4

df[four_mask].head()
```

```python
# checking for null values in df
df.isnull().values.any()
df.isnull().sum()  

```

```python
#  renaming the coumn weather to Weather_Condition
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
df.head()
```

```python
# mean visibility
mean_visibility = df['Visibility_km'].mean()
print(mean_visibility)
```

```python
# number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
wind_and_visibility_mask= (df['Visibility_km']== 25) & (df['Wind Speed_km/h'] ==24)
df[wind_and_visibility_mask].head()
```
